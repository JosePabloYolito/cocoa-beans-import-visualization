```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = "/Users/jpyolito/Desktop/Coding_Dojo/DataScienceFundamentals/datos/Global import of cocoa beans.xlsx"
df = pd.read_excel(filename)
df
```

```python
plt.style.use('Solarize_Light2')
filtro1 = df["Region"] == "Asia & Oceania"
AyO = df.loc[filtro1,:]
lista_year = list(AyO.Year)
lista_imp = list(AyO["Import in tonnes"])
plt.plot(lista_year,lista_imp, marker = "o", c = "chocolate", linestyle = "--", label = "Asia & Oceania")


filtro2 = df["Region"] == "Americas"
America = df.loc[filtro2,:]
lista_year = list(America.Year)
lista_imp = list(America["Import in tonnes"])
plt.plot(lista_year,lista_imp, marker = "s", c = "royalblue", linestyle = "dotted", label = "America")

filtro3 = df["Region"] == "Europe"
Europe = df.loc[filtro3,:]
lista_year = list(Europe.Year)
lista_imp = list(Europe["Import in tonnes"])
plt.plot(lista_year,lista_imp, marker = "^", c = "darkviolet", linestyle = "dashdot", label = "Europe")
plt.title("Import of cocoa beans by region", c = "darkgoldenrod")
plt.xlabel("Year")
plt.ylabel("Tonnes of cocoa beans")
plt.ticklabel_format(axis='y', style='plain') #quita notación cientifica
plt.xticks(range(2015,2020,1))

plt.grid(axis = "y", c = 'brown', 
         alpha = .9, # alpha es para el nivel de transparencia
         linestyle = "dotted")

leg = plt.legend(loc = (1.02,0))
for text in leg.get_texts():
    text.set_color('black')
```
